```python
#Importing Libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#reading data and parsing
r = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup = BeautifulSoup(r.text,'html.parser')
#Extracting data and dataframe creation
rows = soup.find_all('tr')

df = []
for row in rows[1:11]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols[1:]]
    df.append(cols)

df_1 = pd.DataFrame(df, columns=['Country', 'Matches', 'Points', 'Rating'])
df_1['Country'] = df_1.Country.str.split('\n').str[0]

r = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup = BeautifulSoup(r.text,'html.parser')

rows = soup.find_all('tr')

df = []
for row in rows[1:11]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols[1:4]]
    df.append(cols)

df_2 = pd.DataFrame(df, columns=['Player', 'Team', 'Rating'])

r = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
soup = BeautifulSoup(r.text,'html.parser')

rows = soup.find_all('tr')

df = []
for row in rows[1:11]:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols[1:4]]
    df.append(cols)

df_3 = pd.DataFrame(df, columns=['Player', 'Team', 'Rating'])
```
